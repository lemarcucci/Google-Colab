#Calculadora

```python
numero1 = input("Digite o primeiro número ") #pede para o usuário digitar o primeiro número
operacao = input("Digite o número da operção que deseja realizar\n As opções são:\n 1 - Adição\n 2 - Subtração\n 3 - Multiplicação\n 4 - Divisão\n 5 - Calcular o resto da divisão\n ") #pede para o usuário digitar a operação matemática que deseja realizar
numero2 = input("Digite o segundo número ") #pede para o usuário digitar o segundo número
```

```python
#BLOCO DA CONVERSÃO DE TIPO DE VARIÁVEL
convert_primeiro = float(numero1) #converte a variável numero1 em número real
convert_segundo = float(numero2) #converte a variável numero2 em número real
convert_operacao = float(operacao) #converte a variável operacao em número real
type(convert_primeiro) #mostra o tipo da variável convert_primeiro
type(convert_segundo) #mostra o tipo da variável convert_segundo
```

```python
#BLOCO DAS OPERAÇÕES
soma = convert_primeiro + convert_segundo #faz a soma dos dois números
print("A soma dos dois números é", soma)
menos1 = convert_primeiro - convert_segundo #faz a subtração do primeiro número com o segundo número
print("A subtração do primeiro com o segundo é", menos1)
menos2 = convert_segundo - convert_primeiro #faz a subtração do segundo número com o primeiro número
print("A subtração do primeiro com o segundo é", menos2)
vezes = convert_primeiro * convert_segundo #faz a multiplicação dos dois números
print("A multiplicação dos dois número é", vezes)
divisao1 = convert_primeiro / convert_segundo #faz a divisão do primeiro número com o segundo número
print("A divisão do primeiro com o segundo é", divisao1)
divisao2 = convert_segundo / convert_primeiro #faz a divisão do segundo número com o primeiro número
print("A divisão do primeiro com o segundo é", divisao2)
resto1 = convert_primeiro % convert_segundo #faz o resto da divisão do primeiro número com o segundo número
print("O resto da divisão do primeiro com o segundo é", resto1)
resto2 = convert_segundo % convert_primeiro #faz o resto da divisão do segundo número com o primeiro número
print("O resto da divisão do primeiro com o segundo é", resto2)
```
